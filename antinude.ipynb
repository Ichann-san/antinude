```python
from fastapi import FastAPI, HTTPException
from fastapi.middleware.cors import CORSMiddleware
from pydantic import BaseModel
import uvicorn
import nest_asyncio
from pyngrok import ngrok
import base64
from io import BytesIO
from PIL import Image
import numpy as np
import tensorflow as tf  # Uncomment jika modelmu pakai TensorFlow/Keras

# 1. Inisialisasi Aplikasi
app = FastAPI(title="Strokeguard NSFW Backend")

# Wajib: Setup CORS (Cross-Origin Resource Sharing)
# Ini ngasih izin Chrome buat nembak API dari luar domain Colab
app.add_middleware(
    CORSMiddleware,
    allow_origins=["*"],
    allow_credentials=True,
    allow_methods=["*"],
    allow_headers=["*"],
)

# 2. Definisi Struktur Data Payload
class ImagePayload(BaseModel):
    image_base64: str

# 3. Pipeline Model ML (Placeholder)
# Load model kamu di sini: model = tf.keras.models.load_model('model_resnet_nsfw.h5')

def predict_image(image: Image.Image) -> str:
    """
    Fungsi ini untuk memproses gambar dan memprediksi dengan model.
    """
    try:
        # Preprocessing sesuai SOP: Resize dan Normalisasi
        img_resized = image.resize((224, 224))
        img_array = np.array(img_resized) / 255.0
        img_array = np.expand_dims(img_array, axis=0) # Tambah dimensi batch (1, 224, 224, 3)

        # --- LOGIKA INFERENSI MODEL (Uncomment saat model siap) ---
        # prediction = model.predict(img_array)[0][0]
        # if prediction > 0.85: # Threshold yang kita sepakati
        #     return "nsfw"
        # else:
        #     return "safe"

        # (Hapus baris di bawah ini nanti. Ini cuma dummy agar API jalan untuk ditest)
        return "safe"

    except Exception as e:
        print(f"Error saat prediksi: {e}")
        return "safe"

# 4. Endpoint Utama
@app.post("/predict")
async def process_screenshot(payload: ImagePayload):
    try:
        # Decode Base64 string jadi file binary, lalu convert ke format gambar RGB
        image_bytes = base64.b64decode(payload.image_base64)
        image = Image.open(BytesIO(image_bytes)).convert("RGB")

        # Masukkan ke pipeline ML
        status = predict_image(image)

        # Kirim response JSON ke Chrome Extension
        return {"status": status}

    except Exception as e:
        # Error handling kalau base64 corrupt
        raise HTTPException(status_code=400, detail=f"Gagal memproses gambar: {str(e)}")

# 5. Konfigurasi Tunnel Ngrok & Server
# --- MASUKKAN TOKEN NGROK KAMU DI SINI ---
NGROK_AUTH_TOKEN = "3CGdcqZHGDzSmrGDBgGhOTRct5L_48JyTjPdu2UxMrAEQ9s81"
ngrok.set_auth_token(NGROK_AUTH_TOKEN)

# Buka terowongan (tunnel) di port 8000
public_url = ngrok.connect(8000).public_url
print("="*60)
print(f"🚀 API JALAN! COPY URL INI KE background.js KAMU:")
print(f"👉 {public_url}/predict")
print("="*60)

# Konfigurasi uvicorn agar kompatibel dengan Colab
config = uvicorn.Config(app, host="0.0.0.0", port=8000)
server = uvicorn.Server(config)

# Jalankan server secara asinkron
await server.serve()
```
